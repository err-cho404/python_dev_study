```python
import pandas as pd

df = pd.read_excel("NewsResult_20240804-20240805.xlsx",sheet_name="sheet",usecols="E",nrows=50)
df.head()
```

```python
df.to_csv("content.txt",index=False,sep="\t")
```

```python
txtfile =open("./content.txt","r",encoding="utf-8") #r=>읽기전용
sliceText = txtfile.read()
sliceText = sliceText.replace("..."," ").replace("?"," ") #...과 ?를 공백으로 처리
txtfile.close()
sliceText
```

```python
strList = sliceText.split()
wordCloudText = '' 
for listword in strList :
    word = listword.strip("'\"''""") # '와 "를 제거해줌
    wordCloudText = wordCloudText + " " + word
wordCloudText
```
