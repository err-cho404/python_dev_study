```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt

mpl.rcParams['font.family'] = 'Malgun Gothic' # AppleGothic ==> 맥에서는 맑은 고딕대신 사용
mpl.rcParams['axes.unicode_minus'] = False
mpl.rcParams['font.size'] = 10

#삼전도 그래프
df=pd.read_excel('score.xlsx')
plt.scatter(df.영어,df.수학,s=[20,40,60,80,100,120,140,180],marker='^',c=range(len(df)),cmap='viridis')
plt.xlabel('영어점수',color='coral')
plt.ylabel('수학점수',color='blue',alpha=0.5)
```

```python
ramdomdata=np.random.rand(8)*1000
print(ramdomdata,"     ",type(ramdomdata))
print([1,2,3,4,5]*5,"=====>파이썬에서의 리스트 곱")
print(ramdomdata*5,"=====>넌파이의 8행 1열의 1차원 배열에서의 곱")
df=pd.read_excel('score.xlsx')
plt.scatter(df.영어,df.수학,s=ramdomdata,marker='^',c=range(len(df)),cmap='viridis',alpha=0.7)
plt.xlabel('영어점수',color='coral')
plt.ylabel('수학점수',color='blue',alpha=0.5)
```
