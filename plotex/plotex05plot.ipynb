```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt

mpl.rcParams['font.family'] = 'Malgun Gothic' # AppleGothic ==> 맥에서는 맑은 고딕대신 사용
mpl.rcParams['axes.unicode_minus'] = False
mpl.rcParams['font.size'] = 10
df = pd.read_excel('./sample.xlsx',index_col='days')

linecolors=['aqua','coral','gold','grey','lavender','lightblue','lightgreen','orange','pink','teal','yellow','tomato','tan','black','blue']

plt.title('일별 영업사원 판매 현황')
input_data=input("영업사원 명단을 띄어쓰기로 구분하여 입력하세요").split() #화면에서 체크박스로 입력을 받는다.
days = list(df.index) # x축
yname=list(df.columns)

#텍스트 함수 정의
def plot_txt(x,y) : 
    for idx,txt in enumerate(y):
        plt.text(x[idx]-0.02,y[idx]+0.2,txt, color='black')
#함수의 끝

for idx,columns in enumerate(input_data) :
    plt.plot(days,list(df[columns]),color=linecolors[idx])
    plot_txt(days,list(df[columns]))
             
plt.legend(yname,loc = (1.02,0.05))
```
