```python
import matplotlib.pyplot as plt
import matplotlib as mpl #폰트를 설정하기 위한 import
mpl.rcParams['font.family'] = 'Malgun Gothic' # AppleGothic ==> 맥에서는 맑은 고딕대신 사용
mpl.rcParams['axes.unicode_minus'] = False
mpl.rcParams['font.size'] = 10
x=[3,4,5]
y=[6,8,12]
#plt.set_label() => 확인 필요8
#plt.plot([3,4,5],[6,8,12],label = '범례',linewidth=2, marker='o', color = 'c' ,linestyle = ':', ms = 10, mec ='red',mfc='yellow')
plt.plot(x,y,'co-.',alpha=0.2,label = '범례')
#plt.figure(dpi=200)
plt.title('꺽은선 그래프',fontdict={'family' :'HYMyeongJo-Extra','size' : 20, 'color' : 'teal'})
plt.xlabel('X축 데이터',loc='right',fontdict={'color' :'salmon'})
plt.ylabel('Y축 데이터',loc='top',fontdict={'color' :'lightblue'})
plt.xticks([3,4,5],color ='salmon')
plt.yticks(range(6,14,1),color = 'lightblue')
plt.legend(loc = (1.02,0.5)) #'best'는 최적의 차리를 찾아서 배치(기본값), 'upper right' 는 오른쪽 위
for idx,txt in enumerate(y):
    print(idx,'     ',txt)
    plt.text(x[idx]-0.02,y[idx]+0.15,txt, color='pink')

#생성된 그래프 파일로 저장
plt.savefig('linegraph.png',dpi=200)
#plt.show()
```
