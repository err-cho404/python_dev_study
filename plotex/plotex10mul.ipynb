```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt

mpl.rcParams['font.family'] = 'Malgun Gothic' # AppleGothic ==> 맥에서는 맑은 고딕대신 사용
mpl.rcParams['axes.unicode_minus'] = False
mpl.rcParams['font.size'] = 10

df=pd.read_excel('score.xlsx')
df
```

```python
fig,axs = plt.subplots(2,2,figsize=(15,10))#2행2열
fig.suptitle('Multigraph',size=25,color='blue')
axs[0,0].bar(df['이름'],df['국어'],label='국어',color='orange')
axs[0,0].set_title('첫번째 그래프 제목')
axs[0,0].legend(loc='center')
axs[0,0].set(xlabel='이름',ylabel='점수')
axs[0,0].set_facecolor('lightblue')
axs[0,0].grid(linestyle='--',linewidth=1)

axs[0,1].plot(df['이름'],df['국어'],label='국어')
axs[0,1].plot(df['이름'],df['영어'],label='영어')
axs[0,1].plot(df['이름'],df['수학'],label='수학')
axs[0,1].legend(loc=(1.01,0.5))

result=df.groupby(df.학교).size()
y= result.index
x = result.values
axs[1,0].pie(x,labels=y)

axs[1,1].scatter(df['수학'],df['국어'])

```
