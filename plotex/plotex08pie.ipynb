```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt

mpl.rcParams['font.family'] = 'Malgun Gothic' # AppleGothic ==> 맥에서는 맑은 고딕대신 사용
mpl.rcParams['axes.unicode_minus'] = False
mpl.rcParams['font.size'] = 10

values = [2,3,4,5]
#plt.pie(values,labels='TV 냉장고 세탁기 컴퓨터'.split(),startangle=90,autopct='%.2f%%')
colors = plt.get_cmap('Blues')(np.linspace(0.2, 0.7, len(values)))
plt.pie(values,labels='TV 냉장고 세탁기 컴퓨터'.split(),autopct='%.2f%%',explode=[0,0,0,0.1],wedgeprops={'width':0.5},colors=colors)
plt.legend(loc=(0.01,0.01))
```
