```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt

mpl.rcParams['font.family'] = 'Malgun Gothic' # AppleGothic ==> 맥에서는 맑은 고딕대신 사용
mpl.rcParams['axes.unicode_minus'] = False
mpl.rcParams['font.size'] = 10

df=pd.read_excel('./score.xlsx')
df
```

```python
#누적막대그래프 만들기
plt.bar(df['이름'],df.국어,label='국어', color='pink')
plt.bar(df['이름'],df.영어,bottom=df.국어,label='영어',color='lightblue')
plt.bar(df['이름'],df.수학,bottom=df.국어+df.영어,label='수학',color='lavender')
plt.bar(df['이름'],df.과학,bottom=df.국어+df.영어+df.수학,label='과학',color='lightyellow')
plt.bar(df['이름'],df.사회,bottom=df.국어+df.영어+df.수학+df.과학,label='사회',color='wheat')
plt.legend()
```

```python
#누적막대그래프 만들기
index = np.arange(len(df.index))
w=0.15
ws=0.15
plt.bar(index-w,df.국어,label='국어',color='pink',width=w)
plt.bar(index,df.영어,label='영어',color='lightblue',width=w)
plt.bar(index+w,df.수학,label='수학',color='lavender',width=w)
plt.bar(index+(w*2),df.과학,label='과학',color='lightyellow',width=w)
plt.bar(index+(w*3),df.사회,label='사회',color='wheat',width=w)
plt.xticks(index,df['이름'],rotation=30)
plt.legend(ncol=1,loc=(1.01,0.5));
```
