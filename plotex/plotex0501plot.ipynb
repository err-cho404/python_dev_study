```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt

mpl.rcParams['font.family'] = 'Malgun Gothic' # AppleGothic ==>    
mpl.rcParams['axes.unicode_minus'] = False
mpl.rcParams['font.size'] = 10
df = pd.read_excel('./sample.xlsx',index_col='days')

linecolors=['aqua','coral','gold','grey','lavender','lightblue','lightgreen','orange','pink','teal','yellow','tomato','tan','black','blue']

plt.title('   ')

days = df.index # x
yname = list(df.columns)

#  
def plot_txt(x,y) : 
    for idx,txt in enumerate(y):
        plt.text(x[idx]-0.02,y[idx]+0.2,txt, color='black',fontdict={'size':6})
# 

for idx,columns in enumerate(yname) :
    plt.plot(days,list(df[columns]),color=linecolors[idx])
    plot_txt(days,list(df[columns]))
             
plt.legend(yname,loc = (1.02,0.05))



# for i in df[range(1,)]
#     df[i]


```
