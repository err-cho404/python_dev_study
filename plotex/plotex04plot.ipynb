```python
import matplotlib.pyplot as plt
import matplotlib as mpl #폰트를 설정하기 위한 import
mpl.rcParams['font.family'] = 'Malgun Gothic' # AppleGothic ==> 맥에서는 맑은 고딕대신 사용
mpl.rcParams['axes.unicode_minus'] = False
mpl.rcParams['font.size'] = 10
days = [1,2,3,4,5]
data1=[2,3,5,7,9]
data2=[3,4,6,8,10]
data3=[8,5,3,6,12]
plt.plot(days, data1, 'ro:',label = 'a')
plt.plot(days, data2, 'go-.',label = 'b')
plt.plot(days, data3, 'co--', label = 'c')
plt.title('다중 꺽은선 그래프',color = 'lightblue')

ran = [data1,data2,data3]
for i in ran :
    for idx,txt in enumerate(i):
        plt.text(days[idx]-0.02,i[idx]+0.2,txt, color='black')

# def plot_txt(x,y) : 
#     for idx,txt in enumerate(y):
#         plt.text(x[idx]-0.02,y[idx]+0.2,txt, color='black')
# plot_txt(days,data1)
# plot_txt(days,data2)
# plot_txt(days,data3)

plt.legend()
plt.show()
```
