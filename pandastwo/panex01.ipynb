```python
import pandas as pd

# 리스트를 이용한 데이터 프레임 만들어보기 - 비권장 (직접 만들일 없음, 직접 만들더라도 dictionary를 이용해서 만드는 것을 권장)
df = pd.DataFrame( [[1,2,3,4,5,6], range(11,17),'나일등 김영구 최칠득 채치수 정대만 강백호'.split()],index=list("abc"),columns=list("가나다라마바"))
print(df)
print("-"*50)
s = pd.Series([[1,2,3,4,5,6], list(range(11,17))])
print(s)
print(s[0][2:4],type(s[0]))
print(s[1][5])
print(df.T)
df=df.T  # 다른 변수에 담을지 원본을 바꿀지는 선택
#df_c = df.T
#df_c
#print(df)
#print(df_c)
print("-"*50)
print(df['a']) # 컬럼명이나 숫자로된 디폴트컬럼(실용어는 아님) ==>컬럼명을 사용 select a from df
print(df['a']['나']) # ==> select a from df where row = '나'
print(df['a':'c']) #컬럼이 아닌 행을 의미 => 컬럼명만 반환 =>행의 a ~ c를 의미 ==> select a,c from df
print(df['가':'다'][['b','c']]) # ==> select b,c from df where row in ('가','나','다')
print(df['c']['라'])
print(df[3:4]['c'])
```
